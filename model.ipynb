```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data from the xlsx file and return a dataframe
def import_data(data_path):
    df = pd.read_excel(data_path, sheet_name='Sheet1')
    return df
```

```python
sitting_data_path = 'Data/acc_sensor_data_sitting.xlsx'
walking_data_path = 'Data/acc_sensor_data_walking.xlsx'
```

```python
# call the import_data function to get the dataframes
sitting_df = import_data(sitting_data_path)
walking_df = import_data(walking_data_path)
```

```python
# print the first 5 rows of the walking dataframe
walking_df.head()
```

```python
# print thee first 5 rows of the sitting dataframe
sitting_df.head()
```

```python
# print the number of rows and columns in the walking dataframe
print(walking_df.shape)

# print the number of rows and columns in the sitting dataframe
print(sitting_df.shape)
```

```python
# plot the data from the walking dataframe with timestamp as the x-axis and acceleration as the y-axis

walking_df['Timestamp'] = pd.to_datetime(walking_df['Timestamp'])

# Plotting the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.plot(walking_df['Timestamp'].head(10), walking_df['acc_x'].head(10), label='acc_x')
plt.plot(walking_df['Timestamp'].head(10), walking_df['acc_y'].head(10), label='acc_y')
plt.plot(walking_df['Timestamp'].head(10), walking_df['acc_z'].head(10), label='acc_z')

plt.xlabel('Timestamp')
plt.ylabel('Acceleration')
plt.title('Walking Data - Acceleration over Time')
plt.legend()
plt.show()
```

```python
# plot the data from the walking dataframe with timestamp as the x-axis and acceleration as the y-axis

sitting_df['Timestamp'] = pd.to_datetime(sitting_df['Timestamp'])

# Plotting the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.plot(sitting_df['Timestamp'].head(10), sitting_df['acc_x'].head(10), label='acc_x')
plt.plot(sitting_df['Timestamp'].head(10), sitting_df['acc_y'].head(10), label='acc_y')
plt.plot(sitting_df['Timestamp'].head(10), sitting_df['acc_z'].head(10), label='acc_z')

plt.xlabel('Timestamp')
plt.ylabel('Acceleration')
plt.title('Sitting Data - Acceleration over Time')
plt.legend()
plt.show()
```

```python
# divide the walking dataframe into train and test dataframes with 80% of the data in the train dataframe and 20% of the data in the test dataframe
walking_train_df = walking_df.iloc[:int(0.8 * len(walking_df))]
walking_test_df = walking_df.iloc[int(0.8 * len(walking_df)):]
print('shape walking_train_df: ', walking_train_df.shape)
print('shape walking_test_df: ', walking_test_df.shape)

# divide the sitting dataframe into train and test dataframes with 80% of the data in the train dataframe and 20% of the data in the test dataframe
sitting_train_df = sitting_df.iloc[:int(0.8 * len(sitting_df))]
sitting_test_df = sitting_df.iloc[int(0.8 * len(sitting_df)):]
print('shape sitting_train_df: ', sitting_train_df.shape)
print('shape sitting_test_df: ', sitting_test_df.shape)
```
